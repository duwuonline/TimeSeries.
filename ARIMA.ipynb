```
# Lấy 2  Cột làm biến độc lập
inputs = data[['Gia_MoCua','Gia_DongCua']]
# Lấy 1 Cột làm biến phụ thuộc
targets = data['LoiNhuan_Hay_Khong']
# Lấy ngẫu nhiên 80% dòng để máy học và 20% dòng để kiểm tra
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( inputs, targets, train_size = 0.8, test_size=0.2, random_state=16,)
print ('Train set:', X_train.shape, y_train.shape)
print ('Test set:', X_test.shape, y_test.shape)
```

```
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
# Tạo một mô hình cây quyết định
model = DecisionTreeClassifier()

# Thiết lập các tham số cần tối ưu
parameters = {
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'sqrt', 'log2']
}

# Sử dụng GridSearchCV để tìm kiếm các giá trị tốt nhất cho các tham số
grid_search = GridSearchCV(model, parameters, cv=5)
grid_search.fit(X_train, y_train)
# Huấn luyện mô hình với các tham số tốt nhất
best_model = DecisionTreeClassifier(**grid_search.best_params_)
best_model.fit(X_train, y_train)

# Dự đoán và tính độ chính xác trên tập kiểm tra
prediction_DecisionTree = best_model.predict(X_test)
DecisionTree_Acc = accuracy_score(y_test, prediction_DecisionTree)

print(f'Cây quyết định : {DecisionTree_Acc*100}% Độ Chính Xác')
print("Siêu tham số tốt nhất:", grid_search.best_params_)
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import warnings
# Generate some sample data (replace this with your own time series data)
np.random.seed(42)
data = np.random.rand(100) + np.sin(np.linspace(0, 20, 100))

# Create a pandas Series with the data
# Replace 'data' with your actual time series data
time_series = pd.Series(data)

# Define the ARIMA model (ARIMA(5, 0, 1))
p, d, q = 5, 0, 1
P, D, Q, S = 0, 0, 0, 0  # Set seasonal parameters to 0 to ignore seasonality
seasonal_order = (P, D, Q, S)
model = ARIMA(time_series, order=(p, d, q), seasonal_order=seasonal_order)

# Split the data into training and testing sets (70/30 split)
split_index = int(len(time_series) * 0.7)
train_data = time_series[:split_index]
test_data = time_series[split_index:]

# Fit the model to the training data
results = model.fit()

# Forecast future values for the testing period
forecast_steps = len(test_data)
forecast = results.forecast(steps=forecast_steps)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(test_data, forecast)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Print the results
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)

# Plot the original data and the forecasted values
plt.figure(figsize=(10, 6))
plt.plot(time_series, label='Dữ liệu thực tế')
plt.plot(range(len(train_data), len(train_data) + len(forecast)), forecast, label='Dự đoán', color='red')
plt.plot(range(len(train_data), len(train_data) + len(forecast)), test_data, label='Dữ liệu Test', color='green')
plt.title('Dự đoán khối lượng giao dịch bằng mô hình ARIMA')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(False)
plt.legend()
plt.show()
```

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error
# Generate some sample data (replace this with your own time series data)
np.random.seed(42)
data = np.random.rand(100) + np.sin(np.linspace(0, 20, 100))

# Create a pandas Series with the data
# Replace 'data' with your actual time series data
time_series = pd.Series(data)

# Split the data into training and testing sets (70/30 split)
split_index = int(len(time_series) * 0.7)
train_data = time_series[:split_index]
test_data = time_series[split_index:]

# Normalize the data (LSTM requires scaled inputs)
scaler = MinMaxScaler()
train_data = scaler.fit_transform(train_data.values.reshape(-1, 1))
test_data = scaler.transform(test_data.values.reshape(-1, 1))

# Prepare the data for LSTM
def create_sequences(data, sequence_length):
    X, y = [], []
    for i in range(len(data) - sequence_length):
        X.append(data[i : i + sequence_length])
        y.append(data[i + sequence_length])
    return np.array(X), np.array(y)

sequence_length = 10
X_train, y_train = create_sequences(train_data, sequence_length)
X_test, y_test = create_sequences(test_data, sequence_length)

# Build the LSTM model
model = Sequential()
model.add(LSTM(64, input_shape=(sequence_length, 1)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the LSTM model
epochs = 10
batch_size = 32
model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=1)

# Make predictions on the test data
y_pred = model.predict(X_test)

# Inverse transform the predictions and actual values to their original scale
y_pred = scaler.inverse_transform(y_pred)
y_test = scaler.inverse_transform(y_test)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Print the results
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)


# Plot the original data and the forecasted values
plt.figure(figsize=(10, 6))
plt.plot(time_series, label='Actual Data')
plt.plot(range(len(train_data) + sequence_length, len(train_data) + len(y_pred) + sequence_length), y_pred, label='Predicted', color='red')
plt.plot(range(len(train_data) + sequence_length, len(train_data) + len(y_pred) + sequence_length), y_test, label='Test Data', color='green')
plt.title('Predicting Trading Volume with LSTM Models',color = 'r')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()

```
